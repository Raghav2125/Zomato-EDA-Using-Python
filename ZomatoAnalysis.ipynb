```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('zomato.csv') # check utf
data.head()
```

```python
data.shape
```

```python
data.duplicated().sum()
```

```python
data.columns
```

```python
data.info()
```

```python
data = data.drop(['url','address','phone','menu_item','listed_in(type)','listed_in(city)'],axis=1)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
feature_na = [i for i in data.columns if data[i].isnull().sum()>0]
feature_na
```

```python
for i in feature_na:
  print(f"{i} has {np.round((data[i].isnull().sum()/len(data[i])*100),4)}%null values")
```

```python
data.rate.unique()
```

```python
data.dropna(subset=['rate'],axis=0,inplace=True)
```

```python
def split(x):
  return x.split('/')[0].strip() #strip will remove the spaces
```

```python
data.rate = data.rate.apply(split)
```

```python
data.rate
```

```python
data['rate'].replace(['NEW','-'],0,inplace=True)
```

```python
data['rate'] = data['rate'].astype(float)
```

```python
data['rate']
```

```python
data.info()
```

```python
data['approx_cost(for two people)'].unique()
```

```python
def ac(x):
  return x.replace(',','')
```

```python
data['approx_cost(for two people)'] = data['approx_cost(for two people)'].astype(str).apply(ac)
```

```python
data['approx_cost(for two people)'].astype(float)
```

```python
#calculating avg rating for each
rating = pd.pivot_table(data,index='name',values='rate')
rating
```

```python
rating = rating.sort_values(by='rate',ascending=False)
rating[0:15]
```

```python
plt.figure(figsize=(15,8))
sns.barplot(x=rating[0:20].rate,y=rating[0:20].index,orient="h")
plt.show()
```

```python
sns.set_style('whitegrid')
sns.displot(data['rate'])

plt.show()
```

```python
from scipy.stats import normaltest
DataToTest = data['rate']
stat, p = normaltest(DataToTest)
print('Statistics=%.3f, p=%.3f' % (stat, p))

if p>0.05:
  print('Normal Distribution')
else:
  print('Not a Normal Distribution')
```

```python
voting_rating = data.groupby('name')[['votes']].mean().sort_values('votes',ascending=False)
voting_rating
```

```python
voting_rating['name'] = voting_rating.index
```

```python
voting_rating = voting_rating.reset_index(drop=True)
```

```python
voting_rating = pd.merge(voting_rating,data[['rate','name']])
voting_rating.head()
```

```python
vote_top = voting_rating[voting_rating['votes']>500]
vote_top
```

```python
vote_top = vote_top.groupby('name')[['rate']].mean().sort_values('rate',ascending=False)
vote_top
```

```python
plt.figure(figsize=(25,12),dpi=100)
sns.barplot(x=vote_top['rate'],y=vote_top.index,orient="h")
plt.show()
```

```python
bad_count = voting_rating[(voting_rating['rate']<3) & (voting_rating['rate'] > 0)]
bad_count
```
